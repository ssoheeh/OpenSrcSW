```python
import csv
import matplotlib.pyplot as plt
f=open('seoul.csv')
data=csv.reader(f)
next(data)
max_result=[]
min_result=[]
for row in data:
    if row[-1] != '' and row[-2]!='':
        if row[0].split('-')[1]=='12' and row[0].split('-')[2]=='05':
            max_result.append(float(row[-1])) #최고기온 저장
            min_result.append(float(row[-2])) #최저기온 저장 
f.close()
plt.plot(max_result,'deeppink')
plt.plot(min_result,'blue')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
f=open('seoul.csv')
data=csv.reader(f)
next(data)
jan=[]
aug=[]

for row in data:
    month=row[0].split('-')[1]
    if row[-1] != '':
        if month =='01':   # 1월의 최고기온 저장
            jan.append(float(row[-1]))
        if month =='08':   # 8월의 최고기온 저장
            aug.append(float(row[-1]))
f.close()
plt.hist(jan,bins=100,color='b',label='Jan')
plt.hist(aug,bins=100,color='r',label='Aug')
plt.legend()
plt.show()
```
