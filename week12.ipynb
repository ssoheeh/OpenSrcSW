```python
import csv
f = open('seoul.csv','r',encoding='cp949')
data = csv.reader(f)
next(data)
max_diff = 0
max_diff_date = ''
min_temp = 999
min_temp_date = ''
max_temp = -999
max_temp_date = ''
for row in data:
    if row[3]=='' or row[4]=='':
        row[3]=999
        row[4]=999
    row[3]=float(row[3])
    row[4]=float(row[4])
    temp_diff = row[4]-row[3]
    if temp_diff>max_diff:
        max_diff = temp_diff
        max_diff_date = row[0]
    if min_temp>row[3]:
        min_temp = row[3]
        min_temp_date = row[0]
    if max_temp<row[4]:
        max_temp = row[4]
        max_temp_date = row[0]
f.close()
print("일교차가 가장 큰 날\n",max_diff_date) 
print("최저기온이 가장 낮은 날\n",min_temp_date)
print("최고기온이 가장 높은 날\n",max_temp_date)
```
